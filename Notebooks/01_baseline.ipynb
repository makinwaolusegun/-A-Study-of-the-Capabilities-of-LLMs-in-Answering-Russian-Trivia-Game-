```python
def ask_ollama(model_name, question):
    """
    Sends a question to a local Ollama model and returns ONLY the answer.
    """
    prompt = f"""Answer this trivia question using as few words as possible.
Do NOT explain. Do NOT add punctuation, reasoning, or commentary.
Respond with ONLY the answer itself:

Question: {question}
Answer:"""

    cmd = ["ollama", "run", model_name, prompt]
    try:
        output = subprocess.check_output(cmd, text=True)
        return output.strip()
    except subprocess.CalledProcessError as e:
        return f"Error: {e}"
```

```python
!ollama pull mistral-nemo:12b
```

```python
model = "mistral-nemo:12b"

num_questions = 1000
results = []
total_time=0

for i, row in df.head(num_questions).iterrows():
    q = row['question']
    correct = row['answer']

    start_time = time.time()
    pred = ask_ollama(model, q)
    elapsed_time = time.time() - start_time

    results.append({
        "question": q,
        "correct_answer": correct,
        "model_answer": pred,
        "elapsed_time": elapsed_time
    })

    total_time += elapsed_time
    if (i+1) % 50 == 0:
        print(f"{i+1} questions processed...")

# Save results for analysis
results_df = pd.DataFrame(results)
results_df.to_csv(f"ollama_{model.replace(':','_')}_results.csv", index=False)
print("Results saved!")
```
